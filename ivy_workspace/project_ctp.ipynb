```python
#dropping the columns that store the Normalized information about building where the client lives
MODE_cols = [col for col in df.columns if '_MODE' in col]
df.drop(columns = MODE_cols,inplace=True)
MEDI_cols = [col for col in df.columns if '_MEDI' in col]
df.drop(columns = MEDI_cols,inplace=True)
AVG_cols = [col for col in df.columns if '_AVG' in col]
df.drop(columns = AVG_cols,inplace=True)
```

```python
#Find all the columns starting with name FLAG_DOCUMENT , as these are just FLAGS indicating document submission and can be dropped
flag_document_col=df.filter(regex='^FLAG_DOCUMENT',axis=1)

#drop the FLAG_DOCUMENT Columns the column names are stored in temp Series
df.drop(columns=flag_document_col.columns,inplace=True)
```

```python
#These columns 'FLAG_MOBIL','FLAG_EMP_PHONE','FLAG_WORK_PHONE','FLAG_CONT_MOBILE','FLAG_PHONE','FLAG_EMAIL' 
#are FLAGS which need not be used in analysis so drop them
df.drop(columns=['FLAG_MOBIL',
                           'FLAG_EMP_PHONE',
                           'FLAG_WORK_PHONE',
                           'FLAG_CONT_MOBILE',
                           'FLAG_PHONE',
                           'FLAG_EMAIL'],inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
#Percentage of null values in each column
col_with_nullvalues=df.isnull().sum()/len(df)*100
col_with_nullvalues.sort_values(ascending=False)
```

```python
# List the columns having more than 50% missing values
nullvalues_50=col_with_nullvalues[col_with_nullvalues.values>50.0].sort_values(ascending=False)
nullvalues_50
```

```python
#drop the columns with 50 % or more null values
col_with_nullvalues = list(col_with_nullvalues[col_with_nullvalues.values>=50.0].index)
df.drop(labels=col_with_nullvalues,axis=1,inplace=True)
print(len(col_with_nullvalues))
```

```python
# columns with int64 data types have 2 unique values and are flags
#these columns cannot be used in bivariate analysis as they are not categorical.
#df.select_dtypes('int64').apply(pd.Series.nunique, axis = 0)
```

```python
#Let us drop these columns as they have high null values and also their relevance in the analysis is insignificance, 
#we can may be keep one such column 'AMT_REQ_CREDIT_BUREAU_QRT' which could be helpful.
df.drop(columns = ['AMT_REQ_CREDIT_BUREAU_YEAR',
                             'AMT_REQ_CREDIT_BUREAU_MON',
                             'AMT_REQ_CREDIT_BUREAU_WEEK',
                             'AMT_REQ_CREDIT_BUREAU_DAY',
                             'AMT_REQ_CREDIT_BUREAU_HOUR'],inplace=True)
```

```python

# we need to drop other columns which we dont need
df.drop(columns = ['REG_REGION_NOT_LIVE_REGION',
                             'REG_REGION_NOT_WORK_REGION',
                             'LIVE_REGION_NOT_WORK_REGION',
                             'REG_CITY_NOT_LIVE_CITY',
                             'REG_CITY_NOT_WORK_CITY',
                            'LIVE_CITY_NOT_WORK_CITY'],inplace=True)
```

```python
# get the average annuity 
avg_approved_AMT_ANNUITY=(df[(df['TARGET']==1)])['AMT_ANNUITY'].mean()
avg_refused_GOODS_PRICE =(df[(df['TARGET']==0)])['AMT_ANNUITY'].mean()
print("the average annuity of approved applications is ", avg_approved_AMT_ANNUITY)
print("the average annuity of refused applications is ", avg_refused_GOODS_PRICE)
```

```python
#it seems like we have to fill something into the null values
#some AMT_ANNUITY is empty, I will replace the the average AMT_ANNUITY based on its TARGET

df['AMT_ANNUITY'] = np.where(((df['AMT_ANNUITY']=='nan') & (df['TARGET'] == 1) ), avg_approved_AMT_ANNUITY,df['AMT_ANNUITY'] )
df['AMT_ANNUITY'] = np.where(((df['AMT_ANNUITY']=='nan') & (df['TARGET'] == 0) ), avg_refused_GOODS_PRICE,df['AMT_ANNUITY'] )
```

```python
avg_approve_AMT_GOODS_PRICE=(df[(df['TARGET']==1)])['AMT_GOODS_PRICE'].mean()
avg_refuse_AMT_GOODS_PRICE =(df[(df['TARGET']==0)])['AMT_GOODS_PRICE'].mean()
print("the average annuity of approved applications is ", avg_approve_AMT_GOODS_PRICE)
print("the average annuity of refused applications is ", avg_refuse_AMT_GOODS_PRICE)
```

```python
#some AMT_GOODS_PRICE is empty, I will replace the the average AMT_ANNUITY based on its TARGET

df['AMT_GOODS_PRICE'] = np.where(((df['AMT_ANNUITY']=='nan') & (df['TARGET'] == 1) ), avg_approve_AMT_GOODS_PRICE,df['AMT_ANNUITY'] )
df['AMT_GOODS_PRICE'] = np.where(((df['AMT_ANNUITY']=='nan') & (df['TARGET'] == 0) ), avg_refuse_AMT_GOODS_PRICE,df['AMT_ANNUITY'] )

```
